##### 3. Задание
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
# Гипербола
xmin = -20
xmax = 20
dx = 0.1
x = np.around(np.arange(xmin, xmax, dx), decimals=4)
y = 1 / x
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Эллипс
from math import pi
x=1.     #x-position of the center
y=0.5    #y-position of the center
a=2.     #radius on the x-axis
b=1.5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
plt.plot( x+a*np.cos(t) , y+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Окружность
from math import pi

t = np.linspace(0, 2*pi, 100)
plt.plot(np.cos(t), np.sin(t))
plt.axes().set_aspect(1)
plt.grid(color='lightgray',linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##### 5. Задание
Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X1, Y1 = np.meshgrid(X1, Y1)
Z = 2*X + 5*Y
Z1 = (2*X1 + 5*Y1)+50
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X1, Y1, Z1)
show()
```
