1. Решите систему уравнений:
###### y = x**2 – 1
##### exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1 = fsolve(equations, (1, 1))
print (x1, y1)
```

1. Решите систему уравнений и неравенств:
###### y = x**2 – 1
##### exp(x) + x∙(1 – y) > 1

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1 = fsolve(equations, (8, 3))
if x1 > 1:
    print (x1)
if y1 > 1:
    print (y1)
```
