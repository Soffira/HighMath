Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 150)
plt.plot(x, 1*np.cos(x-4)+8)
plt.plot(x, 8*np.cos(x-4.5)+1)
plt.plot(x, 4*np.cos(x-4)+4)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
